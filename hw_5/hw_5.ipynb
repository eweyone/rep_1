# 3.2 Для каждого столбца определите оптимальный с точки зрения потребления памяти тип данных - напишите свои выводы в комментариях

```python
df_pd_1.info() # Цену преобразуем в int32, т.к. должны укладываться, зона примет int16, а все остальные int64 значения примут int8, т.к. должно хватать объема памяти.
               # Все значения object можем преобразовать в категориальный тип category, т.к. уникальных значений в них буквально пара штук.
```

# 3.3 Поменяйте типы данных столбцов датафрейма на выбранные вами в прошлом пункте и сравните потребление памяти до и после оптимизации

```python
df_pd_1 = df_pd_1.fillna(0)

df_pd_1['price']=df_pd_1['price'].astype(np.int32)
df_pd_1['area']=df_pd_1['area'].astype(np.int16)
df_pd_1['bedrooms']=df_pd_1['bedrooms'].astype(np.int8)           # Уже скоратили потребление памяти примерно в 1.5 раза.
df_pd_1['bathrooms']=df_pd_1['bathrooms'].astype(np.int8)
df_pd_1['stories']=df_pd_1['stories'].astype(np.int8)
df_pd_1['parking']=df_pd_1['parking'].astype(np.int8)
df_pd_1.info()
```

```python
df_pd_1['mainroad']=df_pd_1['mainroad'].astype('category')
df_pd_1['guestroom']=df_pd_1['guestroom'].astype('category')
df_pd_1['basement']=df_pd_1['basement'].astype('category')                      # Объем потребления памяти был сокращен до 10кб, что в 5 раз меньше исходника.
df_pd_1['hotwaterheating']=df_pd_1['hotwaterheating'].astype('category')
df_pd_1['airconditioning']=df_pd_1['airconditioning'].astype('category')
df_pd_1['prefarea']=df_pd_1['prefarea'].astype('category')
df_pd_1['furnishingstatus']=df_pd_1['furnishingstatus'].astype('category')
df_pd_1.info()
```

***Потребление памяти до оптимизации составляло 55кб.***

***После оптимизации - 10кб.***
